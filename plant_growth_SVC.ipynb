```python
columns_to_impute=['Soil_Type','Water_Frequency','Fertilizer_Type']
for i in columns_to_impute:
    print("how many unique value for ",i," : ",df[i].unique())
```

```python
df['Soil_Type'] = df['Soil_Type'].map({'loam':0,"sandy":1,"clay":2})
df['Water_Frequency'] = df['Water_Frequency'].map({'bi-weekly':0,"weekly":1,"daily":2})
df['Fertilizer_Type'] = df['Fertilizer_Type'].map({'chemical':0,"organic":1,"none":2})
```

```python
df1 = df.copy()
x = df1.drop("Growth_Milestone", axis=1)  # drop column
y = df1["Growth_Milestone"]  # select column
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python

param={'C':[1],'gamma':[1],'kernel':['linear']}
svc=SVC()
```

```python
grid=GridSearchCV(svc,param_grid=param,refit=True,cv=5,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_pred4=grid.predict(x_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
