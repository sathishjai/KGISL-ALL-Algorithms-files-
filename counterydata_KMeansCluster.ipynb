```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
df=pd.read_csv("/home/sathish/Downloads/archive/Country-data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
columns_to_impute=["country"]
for i in columns_to_impute:
    print("how many unique value for ",i," : ",df[i].nunique())
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df["country"] = label_encoder.fit_transform(df["country"])
```

```python
df.head()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
           
```

```python
x=df[['child_mort','exports','imports','income','life_expec','total_fer','gdpp']]
```

```python
Kmean=KMeans(n_clusters=3)
Kmean.fit(x)
```

```python
Kmean.cluster_centers_
```

```python
Kmean.labels_
```

```python
sample_test=np.array(['10.3','45.5','58.9','19100','76.8','2.13','12200'])
second_test=sample_test.reshape(1,-1)
print(second_test)
Kmean.predict(second_test)
```
