```python
mode_value = df1["Test_Preparation"].mode()
print(f"Filling missing values in column '{'Test_Preparation'} with mode: {mode_value}")
df1["Test_Preparation"].fillna(mode_value, inplace=True)
```

```python
# Calculate the average math score
average_math_score = df1['Math_Score'].mean()
print(f"Average Math Score: {average_math_score}")
```

```python
# Filter students who received standard lunch
standard_lunch_students = df1[df1['Lunch'] == 'Standard']

# Calculate the median reading score for these students
median_reading_score = standard_lunch_students['Reading_Score'].median()
print(f"Median Reading Score for Standard Lunch Students: {median_reading_score}")
```

```python
# Count the number of students who completed the test preparation course
num_completed_test_prep = df1[df1['Test_Preparation'] == 'Completed'].shape[0]
print(f"Number of Students who Completed Test Preparation: {num_completed_test_prep}")
```

```python
# Create the Total_Score column
df1['Total_Score'] = df1['Math_Score'] + df1['Reading_Score'] + df1['Writing_Score']
print(df1.head(10))
```

```python
df1
```

```python
# Find the student with the highest Total_Score
highest_total_score_student = df1.loc[df1['Total_Score'].idxmax()]
print(f"Student with the Highest Total Score: \n{highest_total_score_student}")
```

```python
# Filter students who scored above 90 in all subjects
high_achievers = df1[(df1['Math_Score'] > 90) & (df1['Reading_Score'] > 90) & (df1['Writing_Score'] > 90)]
print(f"High Achievers:\n{high_achievers}")
```

```python
# Group by Ethnicity and calculate the average math score
average_math_by_ethnicity = df1.groupby('Ethnicity')['Math_Score'].mean()
print(f"Average Math Score by Ethnicity:\n{average_math_by_ethnicity}")
```

```python
# Calculate basic statistical summaries
stat_summary = df1[['Math_Score', 'Reading_Score', 'Writing_Score']].agg(['mean', 'median', 'min', 'max'])
print(f"Statistical Summary:\n{stat_summary}")
```
