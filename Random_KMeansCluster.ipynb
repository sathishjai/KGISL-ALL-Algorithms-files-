```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
x=-2.2*np.random.rand(1000,2)
x1=1+2*np.random.rand(500,2)
x[500:1000,:]=x1
plt.scatter(x[:,0],x[:,1],s=50,c='b')
plt.show()
```

```python
Kmean=KMeans(n_clusters=2)
Kmean.fit(x)
```

```python
Kmean.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1],s=50,c='b')
plt.scatter(-1.08984544, -1.09919476,s=200,c='g',marker='s')
plt.scatter(2.04042304,  2.02957788,s=200,c='r',marker='s')
plt.show()
```

```python
Kmean.labels_
```

```python
sample_test=np.array([2,1])
second_test=sample_test.reshape(1,-1)
print(second_test)
Kmean.predict(second_test)
```

```python
sample_test=np.array([-4,1])
second_test=sample_test.reshape(1,-1)
print(second_test)
Kmean.predict(second_test)
```
