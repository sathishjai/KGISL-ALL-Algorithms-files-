```python
f=df[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','Price']]
```

```python
def remove_outliers_iqr(df, f):
    Q1=df[f].quantile(0.25)
    Q3=df[f].quantile(0.75)
    IQR=Q3-Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df [(df [f] >= lower_bound) & (df[f] <= upper_bound)]
```

```python
df= remove_outliers_iqr(df,'B')
```

```python
df= remove_outliers_iqr(df,'ZN')
```
