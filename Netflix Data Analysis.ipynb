<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Null values in Rating Column</p></div>

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Null values in Duration Column</p></div>

```python
# Count of unique values in the duration column
df["duration"].value_counts().head(3)
```

```python
# Most frequent value in the Column
mode_duration = df["duration"].mode()[0]
mode_duration
```

```python
# Filling in the Null values with most frequent value
df["duration"] = df["duration"].fillna(mode_duration)
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Null values in Director Column</p></div>

```python
# We don't have much data to fill Null values in this column
# So, We are filling in the Null values with "Not Mentioned" 
df["director"] = df["director"].fillna("Not Mentioned")
```

```python
# Verifying the absence of Null values
df["director"].isna().sum()
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Null values in Cast Column</p></div>

```python
# We don't have much data to fill Null values in this column
# So, We are filling in the Null values with "Not Mentioned" 
df["cast"] = df["cast"].fillna("Not Mentioned")
```

```python
# Verifying the absence of Null values
df["cast"].isna().sum()
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Removing unwanted values from Rating Column</p></div>

```python
# Unique values in the rating column
# We can see "74min, 84min, 66min" in the rating column which makes no sense
# So, We have to remove them
df["rating"].unique()
```

```python
# Modifying DataFrame with rows which don't include "min" in rating column
df = df[~df["rating"].str.contains("min")]
```

```python
# Unique values in the rating column after the transformation
df["rating"].unique()
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Converting date_added Column to DateTime Format</p></div>

```python
# Removing (,) commas appearing in the date_added column
# With (,) It will be hard for Pandas to convert the Date to DateTime Format
df["date_added"] = df["date_added"].str.replace(",","")
```

```python
# Changing the date_added column to DateTime Format
df["date_added"] = pd.to_datetime(df["date_added"], format="mixed")
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Extracting Year, Month and Date from date_added Column</p></div>

```python
# Extracting Year from date_added column
df["year"] = df["date_added"].dt.year
```

```python
# Extracting Month Names from date_added column
df["month"] = df["date_added"].dt.month_name()
```

```python
# Extracting Date from date_added column
df["date"] = df["date_added"].dt.day
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Extracting Genre from listed_in Column</p></div>

```python
# Extracting first values from listed_in column as genre
df["genre"] = df["listed_in"].str.split(",").str[0]
```

```python
# Unique genres in the DataFrame
df["genre"].value_counts().head(5)
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Extracting Lead Actor from Cast Column</p></div>

```python
# Modifying DataFrame by including only those Value
# Where cast is not equal to "Not Mentioned"
df = df[df["cast"]!="Not Mentioned"]
```

```python
# Extracting first values from cast column as lead_actor
df["lead_actor"] = df["cast"].str.split(",").str[0]
```

```python
# Unique lead_actor in the DataFrame
df["lead_actor"].value_counts().head(5)
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Dropping Unnecessary Column</p></div>

```python
# Dropping date_added, listed_in and cast column
df.drop(["date_added","listed_in","cast"], axis=1, inplace=True)
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Final DataFrame</p></div>

```python
# Final DataFrame
df.sample(5)
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">All the Movies uploaded on Netflix</p></div>

```python
# DataFrame with type is equal to "Movies"
movies_df = df[df["type"]=="Movie"]
movies_df.head(5)
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Extracting the Runtime of Movies as Minutes</p></div>

```python
# Splitting duration column with spaces
# And, Selecting only the numerical part
# And Finally, Saving it to duration_min column
movies_df = movies_df.copy()
movies_df.loc[:, 'duration_min'] = movies_df["duration"].str.split(" ").str[0].astype(int)
```

```python
# Dropping duration column
movies_df.drop(["duration"], axis=1, inplace=True)
movies_df.head(5)
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">All the TV Shows uploaded on Netflix</p></div>

```python
# DataFrame with type is equal to "TV Shows"
tv_shows_df = df[df["type"]=="TV Show"]
tv_shows_df.head(5)
```

<div align="center" style="background-color:gainsboro; border-left:4px solid #E50914; border-right:4px solid #E50914;"><p style="text-align:center; font-family:Verdana; font-size:20px; color:#E50914; padding:10px;">Extracting the Runtime of TV Shows as Seasons</p></div>

```python
# Splitting duration column with spaces
# And, Selecting only the numerical part
# And Finally, Saving it to duration_season column
tv_shows_df = tv_shows_df.copy()
tv_shows_df.loc[:, "duration_season"] = tv_shows_df["duration"].str.split(" ").str[0].astype(int)
```

```python
# Dropping duration column
tv_shows_df.drop(["duration"], axis=1, inplace=True)
tv_shows_df.head(5)
```
