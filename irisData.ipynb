## support vector machine (svm)

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target','flower_name'],axis='columns')
y=df.target
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict([[5.2,3.1,2.4,1.5]])
```

```python
model.predict([[6.7,3.0,5.2,2.3]])
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

## Training parameter

```python
model_reg=SVC(C=1)
model_reg.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model_reg=SVC(C=7)
model_reg.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model_reg=SVC(C=1)
model_reg.fit(x_train,y_train)
model_reg.score(x_test,y_test)
```

```python
model_g=SVC(gamma=1)
model_g.fit(x_train,y_train)
model_g.score(x_test,y_test)
```

```python
model_kernal=SVC(kernel='linear')
model_kernal.fit(x_train,y_train)
model_kernal.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='linear')
svc.fit(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python

param={'C':[0.1,1,2,10,100,1000],'gamma':[1,0.1,0.001,0.0001,0.00001],'kernel':['rbf','poly','linear','sigmoid']}
svc=SVC()
```

```python
grid=GridSearchCV(svc,param_grid=param,refit=True,cv=5,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_pred4=grid.predict(x_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
