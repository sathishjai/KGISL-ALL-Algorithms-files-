```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=4,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python

param={'C':[0.1,1,2,10,100,1000],'gamma':[1,0.1,0.001,0.0001,0.00001],'kernel':['rbf','poly','linear','sigmoid']}
svc=SVC()
```

```python
grid=GridSearchCV(svc,param_grid=param,refit=True,cv=5,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_pred4=grid.predict(x_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
